```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../data/boston_housing.csv.csv')
data.head()
```

```python
# Check for missing values and handeling them
data.isnull().sum()
```

```python
data.fillna(data.median(), inplace=True)
```

```python
data.isnull().sum()
```

# Detecting outliers using IQR method

```python

numerical_cols = data.select_dtypes(include=[np.number]).columns
numerical_cols
```

```python
outliers_info = {}

#detecting outliers in each numerical columns

for col in numerical_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
outliers_info[col] = {
    'Lower Bound': lower_bound,
    'Upper Bound': upper_bound,
    'Outliers Count': len(outliers),
    'Outliers': outliers[col].tolist()
}

for col, info in outliers_info.items():
    print(f"Column: {col}")
    print(f"  Lower Bound: {info['Lower Bound']:.2f}")
    print(f"  Upper Bound: {info['Upper Bound']:.2f}")
    print(f"  Outliers Count: {info['Outliers Count']}")
    print(f"  Outliers: {info['Outliers'][:5]}...")  # Showieng first 5 outliers
    print()
```

```python
#visualizing dataset with outliers plotting boxplots for each numerical column
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(4, 4, i)  
    sns.boxplot(data=data[col])
    plt.title(col)
    plt.tight_layout()

plt.show()
```

```python
#Replacing the outliers with median value
for col in numerical_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
  
    median = data[col].median()
    data[col] = np.where((data[col] < lower_bound) | (data[col] > upper_bound), median, data[col])

print(data.head()) #prinnting the clean data 
```

```python
#Visualizing afterr replacing outliers
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(4, 4, i)  
    sns.boxplot(data=data[col])
    plt.title(col)
    plt.tight_layout()

plt.show()
```

# Conclusion

The **Boston Housing dataset** provides a comprehensive overview of various factors influencing housing prices across different suburbs in Boston. With **14 features** such as crime rate (`CRIM`), average number of rooms (`RM`), and percentage of lower-status population (`LSTAT`), the dataset offers valuable insights into the relationship between socio-economic, geographic, and environmental factors and median home values (`MEDV`).

## Key Observations:

1. **Missing Values**: The dataset contains missing values (denoted as `NA`) in several columns, including `RM`, `LSTAT`, and `MEDV`. These gaps need to be addressed through imputation or removal to ensure data integrity for analysis.

2. **Feature Skewness**: Some features, such as `LSTAT` and `CRIM`, exhibit skewness, indicating a non-uniform distribution. This may require transformations (e.g., log scaling) to normalize the data for modeling.

3. **Outliers**: Certain features, like `CRIM` and `MEDV`, contain extreme values that could disproportionately influence predictive models. Outlier detection and treatment are essential preprocessing steps.

4. **Correlation Insights**: Features like `RM` (average number of rooms) and `LSTAT` (percentage of lower-status population) show strong correlations with `MEDV`. Positive relationships exist between `RM` and `MEDV`, while negative relationships are evident between `LSTAT` and `MEDV`.

5. **Categorical Variables**: The dataset includes categorical variables like `CHAS` (Charles River dummy variable), which can be leveraged to identify distinct patterns in housing prices near the river.

6. **Target Variable Distribution**: The target variable `MEDV` is slightly right-skewed, suggesting that most homes are priced lower, with a few high-value outliers.

## Recommendations for Analysis:

* **Data Cleaning**: Handle missing values using techniques like median imputation for numerical features or mode imputation for categorical ones.

* **Exploratory Data Analysis (EDA)**: Conduct detailed EDA to identify trends, correlations, and potential multicollinearity among features.

* **Feature Engineering**: Create new features or transform existing ones to improve model performance. For example, binning `LSTAT` into categories or normalizing skewed features.

* **Modeling**: Split the dataset into training and testing sets, and apply regression models (e.g., Linear Regression, Random Forest, Gradient Boosting) to predict `MEDV`.

* **Evaluation**: Use metrics like Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and R-squared to evaluate model performance.
